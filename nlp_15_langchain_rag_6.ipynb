4\. Реализуйте диалоговую вопрос-ответную систему, используя подход "conversation retrieval chain". Продемонстрируйте работу, в процессе диалога получив ответы на следующие вопросы:

- Что такое ВКР?
- Когда должны быть доведены темы ВКР?
- Что должно быть отражено в отзыве руководителя?
- Что такое цель исследования?

Реализуйте процесс взаимодействия с моделью в виде бесконечного цикла с ожиданием ввода пользователя при помощи функции `input`. При вводе слова "exit" диалог завершается.


- [x] Проверено на семинаре

````python
from langchain.chains import ConversationalRetrievalChain
from langchain_core.prompts import ChatPromptTemplate

condense_question_template = """
Учитывая следующий диалог и уточняющий вопрос, переформулируй уточняющий вопрос так, чтобы он был самостоятельным.

История диалога:
{chat_history}
Уточняющий вопрос: {question}
Самостоятельный вопрос:"""

condense_question_prompt = ChatPromptTemplate.from_template(condense_question_template)

qa_template = """
Ты ассистент, помогающий отвечать на вопросы.
Используй приведённые ниже фрагменты текста, чтобы ответить на вопрос.
Если ответа в документах нет, скажи, что не знаешь. Используй максимум три предложения и отвечай кратко.

История диалога:
{chat_history}

Контекст:
{context}

Вопрос: {question}
"""

qa_prompt = ChatPromptTemplate.from_template(qa_template)

convo_qa_chain = ConversationalRetrievalChain.from_llm(
    llm=model,
    retriever=retriever,
    condense_question_prompt=condense_question_prompt,
    combine_docs_chain_kwargs={
        "prompt": qa_prompt,
    },
)

convo_qa_chain(
    {
        "question": "What are autonomous agents?",
        "chat_history": "",
    }
)
````

````python
from langchain_core.messages import HumanMessage, AIMessage

chat_history = []

print("Прим.: для выхода напишите exit.\n\n")

while True:
  question = input("\nВведите вопрос: ")
  if question.strip().lower() == "exit":
    print("\n\nКонец диалога")
    break

  result = convo_qa_chain.invoke({"question": question,"chat_history": chat_history})
  print("Ответ модели:", result['answer'])
  print('='*30)

  #chat_history += f"Вопрос: {question}\n Ответ LLM: {result['answer']}"
  chat_history.append(HumanMessage(content=question))
  chat_history.append(AIMessage(content=result['answer']))
````

````python
import textwrap

from langchain_core.messages import HumanMessage, AIMessage

chat_history = []

print("Прим.: для выхода напишите exit.\n\n")

while True:
  question = input("\nВведите вопрос: ")
  if question.strip().lower() == "exit":
    print("\n\nКонец диалога")
    break

  result = convo_qa_chain.invoke({"question": question,"chat_history": chat_history})
  print("Ответ модели:", textwrap.fill(result['answer'], width=80))
  print('='*30)

  #chat_history += f"Вопрос: {question}\n Ответ LLM: {result['answer']}"
  chat_history.append(HumanMessage(content=question))
  chat_history.append(AIMessage(content=result['answer']))
````
